```python
import torch

class TopKMask(torch.nn.Module):
    def forward(self, x, k):
        # PyTorch handles this fine during training
        values, indices = torch.topk(x, k)
        mask = torch.zeros_like(x).scatter_(1, indices, 1.0)
        return x * mask

model = TopKMask()
dummy_input = torch.randn(1, 10)
# Exporting this might fail or produce weird results depending on the Opset
#torch.onnx.export(model, (dummy_input, 5), "model.onnx")
torch.onnx.export(model, (dummy_input, 5), "model.onnx", opset_version=17)
```
